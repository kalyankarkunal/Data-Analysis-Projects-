```python
# SCALING OR NORMALIZATION - ONLY DONE FOR INPUTS
# SCALING OR NORMALIZATION - scaling down the value of 2 or more different entities
# 1.BEFORE NORMALIZATION/SCALING
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) #random.seed(0) keeps the generated values constant
#using numpy,access the random packages and access the randint lib, in the range of 1-50,generate 30 integers
x1 = np.random.randint(1,50,30)
x1 = np.sort(x1) # sorting up the values in accending order
x2 = np.random.randint(10000,70000,30)
plt.plot(x1,x2)

```

```python
# AFTER NORMALIAZATION
x1min = min(x1)
x1max = max(x1)
x2min = min(x2)
x2max = max(x2)
x1norm = (x1 - x1min)/(x1max - x1min)
x2norm = (x2 - x2min)/(x2max - x2min)
plt.plot(x1norm,x2norm)
```

```python
#LOGISTIC REGRESSION - ITS CLASSIFICATION TECHNIQUE(SUPERVISED LEARNING)
# DATASET- social networks ads - link-'https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Social_Network_Ads.csv'
```

```python
# 1 Gather data and create dataframe
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/Social_Network_Ads.csv')
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.size
```

```python
#PURCHASED COLUMN - 0 AND 1
#0 - NOT PURCHASED
#1- IS PURCHASED
#INPUT - AGE, Estimated salary
#OUTPUT - PURCHASED
```

```python
#TAKING I/P AND O/P
#INPUT IS ALWAYES IN 2 DIMENSIONAL
#Applying slicing method
x = df.iloc[:,2:4].values
x
```

```python
#output is always 1 dimensional
y = df.iloc[:,4].values
y
```

```python
#5 train and test variables
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 0)
#random_state is vary similar to random to random.seed(0)(keeps the generated values constant)
# BY default - 75% trainig data - training variable
# rest 25% is used as testing data
```

```python
#the moment we run the train_test_split
#the data of x variables is divided into x_train(75%) and x_test(25%)
#the data of y variables id divided into y_train(75%) and y_test(25%)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#NORMALIZATION or SCALING
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
# APPLY CLASSIFIER/REGRSSION/CLUSTORER - CLASSIFIAR
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#Fitting the model
model.fit(x_train,y_train) # we are using the x_train and y_train values to trai / fit the model
```

```python
#predictor variable
y_pred = model.predict(x_test)
y_pred #predicted output values
```

```python
y_test
```

```python
# to find out he accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test) * 100
```

```python
#INDIVIASUAL PREDICTION
model.predict([[20,25000]]) #i want know if a person aged 20 and salary 25000 has purchsed thhe product or not
# model.predict([[28,28000]]) 
# model.predict([[80,25000]]) 
```
